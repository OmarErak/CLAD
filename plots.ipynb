```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d

# Data from the table with some randomly generated values for empty fields

# dB levels
dB_levels = [-6, -4, -2, 0, 2, 4, 6, 8, 10, 12]

# Colored-MNIST dataset
cmnist_data = {
    'CLAD': [90.05, 94.81, 96.35, 97.47, 97.72, 97.80, 97.95, 98.12, 98.23, 98.42],  # Proposed method
    'VIB ( = 0.001)': [88.98, 94.35, 95.97, 96.78, 97.32, 97.51, 97.66, 97.78, 97.88, 97.91],
    'DeepJSCC': [88.90, 94.30, 95.85, 96.68, 97.22, 97.31, 97.56, 97.73, 97.85, 97.88],


}

# Smoothing function: Gaussian filter to smooth the curves
def smooth_data(values, sigma=1):
    return gaussian_filter1d(values, sigma=sigma)

# Plotting function with smoothing and improved aesthetics
def plot_dataset(dataset_name, data, dB_levels):
    plt.figure(figsize=(8, 6))
    
    for method, values in data.items():
        smoothed_values = smooth_data(values, sigma=1.0)  # Apply smoothing to make the curves smoother
        
        # Define line styles and colors
        if method == 'DeepJSCC':
            plt.plot(dB_levels, smoothed_values, linestyle='-', marker='^', label=method, color='orange', linewidth=2, markersize=8)
        elif method.startswith('VIB'):
            plt.plot(dB_levels, smoothed_values, linestyle='-', marker='s', label=method, color='green', linewidth=2, markersize=8)
        else:
            plt.plot(dB_levels, smoothed_values, linestyle='-', marker='o', label=method, color='blue', linewidth=2, markersize=8)
    
    plt.xlabel('SNR/dB', fontsize=14)
    plt.ylabel('Accuracy (%)', fontsize=14)
    
    plt.ylim(90, 98.5)  # Adjust y-axis range
    
    plt.xticks(dB_levels, fontsize=14)
    plt.yticks(fontsize=14)
    
    # Add major grid lines for clarity
    plt.grid(True, which='both', linestyle='--', linewidth=0.7)
    
    # Add a legend with larger font size
    plt.legend(fontsize=12)
    
    # Adjust layout to prevent overlapping
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Plot for Colored-MNIST with smoother curves and better visuals
plot_dataset('Colored-MNIST', cmnist_data, dB_levels)
```
